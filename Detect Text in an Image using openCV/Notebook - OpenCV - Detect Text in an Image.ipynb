```python
 import cv2
```

```python
import pytesseract
```

```python
!wget "https://cdn.pixabay.com/photo/2014/09/02/12/24/newspaper-433589_1280.jpg"
```

```python
image = cv2.imread("newspaper-433589_1280.jpg")
```

```python
def cv2_imshow(image):
    cv2.imshow("",image)
    k = cv2.waitKey(0) & 0xFF
    if(k==27):
        cv2.destroyAllWindows()
```

```python
cv2_imshow(image)
```

```python
pytesseract.image_to_string(image)
```

```python

boxes = pytesseract.image_to_boxes(image)
print(boxes)
```

```python
height, width, colors = image.shape
```

```python
color = (255,255,0)
thickness = 2

for box in boxes.splitlines():
    box = box.split(' ')
#     print(box)
    start_x = int(box[1])
    start_y = height - int(box[2])
    end_x = int(box[3])
    end_y = height - int(box[4])
    cv2.rectangle(image, (start_x,start_y), (end_x, end_y), color, thickness)
```

```python
cv2_imshow(image)
```

```python
cv2.imwrite("Letters_outlined.jpg", image)
```

```python
gray_image =  cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_image)
```

```python
threshold = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
```

```python
cv2_imshow(threshold)
```

```python
text = pytesseract.image_to_string(threshold)
print(text)
```

```python
boxes = pytesseract.image_to_boxes(threshold)
print(boxes)
```

```python
height, width = threshold.shape
threshold.shape
```

```python
color = (100,1000,1000)
thickness = 2

for box in boxes.splitlines():
    box = box.split(' ')
#     print(box)
    start_x = int(box[1])
    start_y = height - int(box[2])
    end_x = int(box[3])
    end_y = height - int(box[4])
    cv2.rectangle(threshold, (start_x,start_y), (end_x, end_y), color, thickness)
```

```python
cv2_imshow(threshold)
```

```python
cv2.imwrite("Letters_outlined_with_thresholding.jpg", threshold)
```
